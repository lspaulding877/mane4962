```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import skimage.io
from google.colab import drive

drive.mount('/content/drive')

image = skimage.io.imread("/content/drive/MyDrive/Colab Notebooks/seg2.jpg")
```

```python
# Reshape the image to a 2D array of pixels
pixels = image.reshape((-1, 3))
```

```python
# Create and fit KMeans model
kmeans = KMeans(n_clusters=3)
kmeans.fit(pixels)
```

```python
# Get cluster labels for each pixel
cluster_labels = kmeans.labels_

# Reshape the labels back into the original image shape
segmented_image = cluster_labels.reshape(image.shape[:2])
```

```python
# Display the segmented image
plt.imshow(segmented_image)
plt.show()
```
