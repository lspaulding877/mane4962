```python
from google.colab import drive
from google.colab import files

# Mount Google Drive
drive.mount('/content/drive')
# Go to the current folder
%cd /content/drive/MyDrive/Colab\ Notebooks/


%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
from matplotlib import cm
import itertools

datafile = 'housing_prices.txt'
cols = np.loadtxt(datafile,delimiter=',',usecols=(0,1),unpack=True)
X = np.transpose(np.array(cols[:-1]))
y = np.transpose(np.array(cols[-1:]))
m = y.size
X_plot = np.insert(X,0,1,axis=1)

#figure
plt.figure(figsize=(10,6))
plt.plot(X[:,0],y[:,0],'b+',markersize=12)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Price in $10,000s')
```

```python
import tensorflow as tf
from tensorflow import keras
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from tensorflow.python.keras.optimizers import *
import numpy as np

model = Sequential()
model.add(Dense(5, input_dim=1, activation='relu'))
model.add(Dense(7, activation='relu'))
model.add(Dense(7, activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

model.summary()
```

```python
history = model.fit(X, y, epochs=5, batch_size=1, validation_data=(X, y))
y_160k1 = model.predict(np.array([[16]]))
print('Prediction at 160k, batch = 1:',y_160k1)

lossb1 = history.history['loss']
```

```python
history = model.fit(X, y, epochs=5, batch_size=5, validation_data=(X, y))
y_160k5 = model.predict(np.array([[16]]))
print('Prediction at 160k, batch = 5:',y_160k5)

lossb5 = history.history['loss']
```

```python
history = model.fit(X, y, epochs=5, batch_size=10, validation_data=(X, y))
y_160k10 = model.predict(np.array([[16]]))
print('Prediction at 160k, batch = 10:',y_160k10)

lossb10 = history.history['loss']
```

```python
history = model.fit(X, y, epochs=5, batch_size=20, validation_data=(X, y))
y_160k20 = model.predict(np.array([[16]]))
print('Prediction at 160k, batch = 20:',y_160k20)

lossb20 = history.history['loss']
```

```python
plt.figure(figsize=(10,6))
plt.plot(X[:,0],y[:,0],'b+',markersize=12, label='Training data')
plt.plot(16,y_160k1,'go',markersize=12, label='batch = 1')
plt.plot(16,y_160k5,'ro',markersize=12, label='batch = 5')
plt.plot(16,y_160k10,'co',markersize=12, label='batch = 10')
plt.plot(16,y_160k20,'mo',markersize=12, label='batch = 20')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Price in $10,000s')

plt.legend()
plt.show()
```

```python
plt.plot(lossb1, label='batch=1')
plt.plot(lossb5, label='batch=5')
plt.plot(lossb10, label='batch=10')
plt.plot(lossb20, label='batch=20')
plt.title('Objective function for each batch size (model loss)')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
```

It can be noted that the objective function is very similar for batch size of 5, 10, and 20, but the loss is noticably larger for a batch size of one.  Additionally, the loss seems to decrease as the batch size increases.
