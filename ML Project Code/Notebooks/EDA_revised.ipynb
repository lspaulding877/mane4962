```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive

drive.mount('/content/drive')

from scipy.io import loadmat

data = loadmat('/content/drive/MyDrive/Colab Notebooks/X1.mat')
dt = data['X1']
```

```python
print(dt)
```

```python
dt = pd.DataFrame(dt)
dt.columns = ['x1', 'x_d1', 'alpha1','alpha_d1', 'voltage','time']  # Example column names
print(dt)
```

```python
column_means = dt.mean()
column_medians = dt.median()
column_stds = dt.std()
column_min = dt.min()
column_max = dt.max()
```

```python
descriptive_stats = {
    'mean': column_means,
    'median': column_medians,
    'std': column_stds,
    'min': column_min,
    'max': column_max
}

summary_table = pd.DataFrame(descriptive_stats)
print(summary_table)
```

```python
import matplotlib.pyplot as plt

# Get the column names
x_cols = dt.columns[:4]  # First 4 columns
y_col = dt.columns[-1]   # Last column

# Create the plot
fig, ax = plt.subplots(figsize=(10, 8))  # Create a single plot

for x_col in x_cols:
    ax.plot(dt[y_col], dt[x_col], label=x_col)  # Plot each column against the last column

ax.set_xlabel('Time (s)')  # Set x-axis label
ax.set_ylabel('States')  # Set y-axis label
ax.set_title('States vs. Time')  # Set plot title
ax.legend()  # Add a legend to distinguish the lines

plt.tight_layout()  # Adjust spacing
plt.show()  # Display the plot
```
